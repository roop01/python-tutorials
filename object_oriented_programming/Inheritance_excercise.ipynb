**Excercise -1**

An apparel shop wants to manage the items which it sells.

**Class Description:**

**Apparel class:**

1. Initialize static variable counter to 100.

2. In the constructor, auto-generate item_id starting from 101 prefixed by "C" for cotton apparels and "S" for silk apparels.
Example – C101, S102, S103, C104 etc.
3. calculate_price(): Add 5% service tax on the price of the apparel and update attribute, price with the new value.

**Cotton class:**
1. While invoking parent constructor from child constructor, pass "Cotton" as item_type
2. calculate_price(): Update attribute, price of Apparel class based on rules given below-

    Add service tax on price by invoking appropriate method of Apparel class

    Apply discount on price

    Add 5% VAT on final price

**Silk class:**
1. While invoking parent constructor from child constructor, pass "Silk" as item_type.
2. calculate_price(): Update attribute, price of Apparel class based on rules given below. -

    Add service tax on price by invoking appropriate method of Apparel class

    Identify points earned based on rules given below:
        Silk apparels with price more than Rs. 10000, earn 10 points and anything less than or equal to that earn 3 points
    Initialize attribute, points with the identified points
    Add 10% VAT on price

Note: Perform case sensitive string comparison  

For testing:
1. Create objects of Cotton class and Silk class
2. Invoke calculate_price() on Cotton objects and Silk objects
3. Display their details

```python
class Apparel:
    counter=100
    def __init__(self, price, item_type):
        Apparel.counter += 1
        self.__item_type = item_type
        self.__price = price
        if item_type.lower() == "cotton":
            self.__item_id = "C" + str(Apparel.counter)
        if item_type.lower() == "silk":
            self.__item_id = "S" + str(Apparel.counter)

    def calculate_price(self):
        self.__price = self.__price + (self.__price * 0.05)

    def get_item_id(self):
        return self.__item_id

    def get_item_type(self):
        return self.__item_type

    def get_price(self):
        return self.__price

    def set_price(self, price):
        self.__price = price

class Cotton(Apparel):
    def __init__(self, price, discount):
        super().__init__(price, "Cotton")
        self.__discount = discount

    def calculate_price(self):
        super().calculate_price()
        price = self.get_price()
        price = price - (price * self.__discount/100)
        price = price + (price * 0.05)
        self.set_price(price)

    def get_discount(self):
        return self.__discount

class Silk(Apparel):
    def __init__(self, price):
        super().__init__(price, "Silk")
        self.__points = 0

    def calculate_price(self):
        super().calculate_price()
        price = self.get_price()
        if price >10000:
            self.__points = 10
        else:
            self.__points = 3
        price = price + (price * 0.1)
        self.set_price(price)

    def get_points(self):
        return self.__points

c1= Cotton(1000,10)
c1.calculate_price()
print(c1.get_item_id())
print(c1.get_item_type())
print(c1.get_price())

s1= Silk(10000)
s1.calculate_price()
print(s1.get_item_id())
print(s1.get_item_type())
print(s1.get_price())
print(s1.get_points())



```

**Excercise 2.**

Spice Hut is a tiffin service provider which home delivers dinner to their customers – Occasional customers and Regular customers.
Write a Python program to implement class.

**OccasionalCustomer class:**
1. Initialize static variable counter to 1000
2. Inside the constructor, auto-generate bill_id starting from 1001 prefixed by "O"
3. validate_distance_in_kms():

    Delivery distance in kms should be between 1 and 5 (both inclusive).
    If so, return true. Else, return false

        Distance in kms                                 Delivery charge in Rs.
        Between 1 and 2(both inclusive)                    Rs. 5 per km
        Between 2 and 5(excluding 2,including 5)           Rs. 7.5 per km

4. calculate_bill_amount():
      
      Validate distance in kms. If valid, compute bill amount based on details mentioned below.
          -> Occasional customers can order only one tiffin per person
          -> Cost/tiffin is Rs. 50
          -> Delivery charges based on distance is as mentioned in the table
          -> Total bill amount includes tiffin cost and delivery charge
          -> Set attribute, bill_amount with the computed bill amount value and return it.
          -> If invalid, set attribute, bill_amount as -1 and return it

**RegularCustomer class:**

1. Initialize static variable counter to 1000
2. Inside the constructor, auto-generate bill_id starting from 1001 prefixed by "R"
3. validate_no_of_tiffin(): Validate number of tiffins.
    No.Of.tiffin should be between 1 and 7 (both inclusive).

    If value of no_of_tiffins is valid, return true. Else, return false

4. calculate_bill_amount():

    Validate number of tiffins. If valid, compute bill amount based on details mentioned below-
        -> Regular customers ger free delivery charge
        -> Cost per tiffin is Rs.50
        -> Total cost for no.of.tiffin ordered is calculated based on the below table
              No.Of.Tiffin	          Tiffin Cost in Rs.
              Between 1 and 2    (both inclusive)	Rs.50 per tiffin
              Between 3 and 5    (both inclusive)	10% discount on total cost
              Between 6 and 7    (both inclusive)	30% discount on total cost

        -> Set the attribute 'bill_amount' with the computed value of total cost and return it

Note: Perform case sensitive string comparison

For testing:
Create objects of OccasionalCustomer and RegularCustomer classes
Invoke calculate_bill_amount() on OccasionalCustomer and RegularCustomer objects
Display the details of the customer.

```python
from abc import ABCMeta,abstractmethod
class Customer(metaclass=ABCMeta):
    def __init__(self,customer_name):
        self.__customer_name=customer_name
        self.bill_id = None
        self.bill_amount = 0

    def get_customer_name(self):
        return self.__customer_name

    @abstractmethod
    def calculate_bill_amount(self):
        pass

class OccasionalCustomer(Customer):
    __counter = 1000
    def __init__(self,customer_name,distance_in_kms):
        super().__init__(customer_name)
        self.__distance_in_kms = distance_in_kms
        OccasionalCustomer.__counter += 1
        self.__bill_id = "O" + str(OccasionalCustomer.__counter)

    def validate_distance_in_kms(self):
        if self.__distance_in_kms >= 1 and self.__distance_in_kms <= 5:
            return True
        else:
            return False

    def calculate_bill_amount(self):
        delivery_charge = 0
        if self.validate_distance_in_kms():
            if self.__distance_in_kms <= 2:
                delivery_charge = 5 * self.__distance_in_kms
            else:
                delivery_charge = 7.5 * self.__distance_in_kms
            self.bill_amount = 50 + delivery_charge
            return self.bill_amount
        else:
            self.bill_amount = -1
            return self.bill_amount

    def get_distance_in_kms(self):
        return self.__distance_in_kms

    def get_bill_id(self):
        return self.__bill_id

class RegularCustomer(Customer):
    __counter = 1000
    def __init__(self,customer_name,no_of_tiffins):
        super().__init__(customer_name)
        RegularCustomer.__counter += 1
        self.__bill_id = "R" + str(RegularCustomer.__counter)
        self.__no_of_tiffins = no_of_tiffins

    def validate_no_of_tiffin(self):
        if self.__no_of_tiffins >= 1 and self.__no_of_tiffins <= 7:
            return True
        else:
            return False

    def get_no_of_tiffins(self):
        return self.__no_of_tiffins

    def calculate_bill_amount(self):
        if self.validate_no_of_tiffin():
            self.bill_amount = 50*self.__no_of_tiffins
            if self.__no_of_tiffins >= 2 and self.__no_of_tiffins <= 5:
                return self.bill_amount*.90
            elif self.__no_of_tiffins >= 5 and self.__no_of_tiffins <= 7:
                return self.bill_amount*.70
            else:
                return self.bill_amount
        else:
            self.bill_amount = -1
            return self.bill_amount

    def get_bill_id(self):
        return self.__bill_id


o1 = OccasionalCustomer("Alwyn",3)
print(o1.calculate_bill_amount())
print(o1.get_bill_id())
print(o1.get_customer_name())
print(o1.get_distance_in_kms())

r1 = RegularCustomer("Rupali",3)
print(r1.calculate_bill_amount())
print(r1.get_bill_id())
print(r1.get_customer_name())
print(r1.get_no_of_tiffins())



```
